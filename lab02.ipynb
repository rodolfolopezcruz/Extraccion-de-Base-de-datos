```python
import pandas as pd
```

```python
cantidad = 1000000
```

```python
import random 
```

```python
influencers = ["Ana","Luis","Martha","Carolina","Sofia","Pedro","Elena","Diego"]
tipos_post = ["Playa","Viaje","Comida","Hotel","Selfie","Atardecer"]
```

```python
dfMillon = pd.DataFrame({
    "influencer" : [random.choice(influencers) for _ in range(cantidad)],
    "likes" : [random.randint(1, 50000) for _ in range(cantidad)],
    "comentarios" : [random.randint(1, 5000) for _ in range(cantidad)],
    "fecha" : [random.randint(2020, 2026) for _ in range(cantidad)],
    "tipo_post" : [random.choice(tipos_post) for _ in range(cantidad)]
})
```
